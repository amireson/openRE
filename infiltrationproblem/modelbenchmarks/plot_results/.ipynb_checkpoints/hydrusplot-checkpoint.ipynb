```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
def RMSEfun(x,y):
    return np.sqrt(np.mean((x-y)**2))

def GetError(df):
    dt=df['dt']
    Qin=df['Qin'].values
    Qout=df['Qout'].values
    S=df['S'].values
    
    dS=S[-1]-S[0]
    dQ=np.sum((Qin-Qout)*dt)
    
    QIN=np.sum(Qin)*dt
    QOUT=np.sum(Qout)*dt
    err=1-dQ/dS
    
    MBerr=(dS-dQ)
    RMSE=np.sqrt(np.mean((np.diff(S)-(Qin[1:]-Qout[1:])*dt[1:])**2))
    MBabserr=np.sum(np.abs(np.diff(S)-(Qin[1:]-Qout[1:])*dt[1:]))
    return MBerr, RMSE
```

```python
def GetIresonData(mypath):
    fname='../%s/WB.npy'%mypath
    data=np.load(fname)
    t=pd.date_range(start='1999-10-1',freq='D',periods=len(data))
    WB=pd.DataFrame(index=t)
    S=data[:,0]*1000
    Qin=data[:,1]*1000
    Qout=data[:,2]*1000
    
    dt=1.
    df=pd.DataFrame(index=t)
    df['dt']=dt
    df['Qin']=Qin
    df['Qout']=Qout
    df['S']=S
    df['dS']=np.hstack([0,np.diff(S)])
    df['cQ']=np.cumsum((Qin-Qout)*dt)
    return df
```

```python
def GetHydrusData(mypath):
    WB=pd.read_csv('../%s/T_Level.out'%mypath,skiprows=[0,1,2,3,4,5,7],skipfooter=1,delim_whitespace=True,index_col=0)
    WB.index.name=None
    Qin=-WB['vTop'].values*1000
    Qout=-WB['vBot'].values*1000
    S=WB['Volume'].values*1000
    start='1999-10-1'
    WB['t']=WB.index
    t=pd.to_datetime(start)+WB['t'].apply(pd.Timedelta, unit='D')
    dt=np.hstack([0,np.diff(WB.index)])
    
    df=pd.DataFrame(index=t)
    df['dt']=dt
    df['Qin']=Qin
    df['Qout']=Qout
    df['S']=S
    df['dS']=np.hstack([0,np.diff(S)])
    df['cQ']=np.cumsum((Qin-Qout)*dt)
    return df
```

```python
ire=GetIresonData('IresonRun')
ire2=GetIresonData('IresonOptRun')
hyd=GetHydrusData('InfiltrationProblem')
hyd2=GetHydrusData('InfiltrationProblem_HighAc')
```

```python
print('%.4f,  %.2e'%GetError(ire))
print('%.4f,  %.2e'%GetError(ire2))
print('%.4f,  %.2e'%GetError(hyd))
print('%.4f,  %.2e'%GetError(hyd2))
```
